## Problem 1 — Write and Read a Text File *(10 points)*

Create a text file with Python, confirm that it exists, and read its contents.

### What to do

1. Import the `os` module.
2. Using a `with open(...)` block in **write mode** (`"w"`), create a file named `shop_notes.txt` containing exactly these three lines:

   ```text
   Eagle Beans daily log
   Lattes sold: 42
   Muffins sold: 17
   ```

   Add `"\n"` after each line so that the lines are stored separately. Use `encoding="utf-8"` when opening the file.
3. Use `os.path.exists("shop_notes.txt")` to confirm that the file exists and print the result.
4. Open the file in **read mode** (`"r"`) and print its full contents.
5. Open it once more, use `.readlines()` to count the lines, and print:

   ```text
   Line count: 3
   ```

6. Add a one-sentence **comment** explaining why `with open(...)` is preferred over manually calling `open()` and `close()`.

### Hints

- Write mode (`"w"`) creates a new file or overwrites an existing file.
- Read mode (`"r"`) reads an existing file.
- `os.path.exists(path)` returns `True` or `False`.
- `.read()` returns the entire file as one string.
- `.readlines()` returns a list in which each item represents one line.


````python
# Meghan Kajihara — Problem 1

import os

shop_notes_file = "shop_notes.txt"

with open(shop_notes_file, "w", encoding="utf-8") as file:
  file.write("Eagle Beans daily log \n")
  file.write("Lattes sold: 42 \n")
  file.write("Muffins sold: 17\n")

print("File exists?", os.path.exists("shop_notes.txt"))

with open(shop_notes_file, "r", encoding="utf-8") as file:
  dailylog = file.read()

print(dailylog)

with open(shop_notes_file, "r", encoding="utf-8") as file:
  line_count = len(file.readlines())

print("Line count:", line_count)

# with open is preferred over manually calling open and close because it closes the file automatically at the end of the block
````

## Problem 2 — Rename and Copy a File *(10 points)*

Rename the file created in Problem 1 and make a backup copy. **Run Problem 1 before this problem.**

### What to do

1. Import `os` and `shutil`.
2. Use `os.rename(...)` to rename `shop_notes.txt` as `shop_log.txt`.
3. Use `os.path.exists(...)` to print the following checks:

   ```text
   Old name exists? False
   New name exists? True
   ```

4. Use `shutil.copy(...)` to create a backup named `shop_log_backup.txt`.
5. Print whether the backup exists.
6. Use `os.listdir(".")` to obtain the current folder contents. From that result, print the filenames `shop_log.txt` and `shop_log_backup.txt` so the grader can confirm that both exist. Other files in the folder do not need to be printed.
7. Add a one-sentence **comment** explaining the difference between renaming a file and copying a file.

### Hints

- `os.rename(old_name, new_name)` changes the filename; the old filename no longer remains.
- `shutil.copy(source, destination)` keeps the source file and creates a duplicate.
- `os.listdir(".")` returns the contents of the current folder.
- The order returned by `os.listdir(...)` is not guaranteed, so do not depend on a particular order.


````python
# Meghan Kajihara — Problem 2

import os
import shutil

old_name = "shop_notes.txt"
new_name = "shop_log.txt"
backup_name = "shop_log_backup.txt"

os.rename(old_name, new_name)

print("Old name exists?", os.path.exists(old_name))
print("New name exists?", os.path.exists(new_name))

shutil.copy(new_name, backup_name)

print("Backup exists?", os.path.exists(backup_name))

for file in os.listdir("."):
  if file.startswith("shop_log"):
    print(file)

# Renaming a file only changes the name, but creating a copy creates a new file with a new name but the same contents.
````

## Problem 3 — Work with a CSV File *(10 points)*

Write tabular data to a CSV file, read it back, and calculate total sales.

### What to do

1. Import the `csv` module.
2. Using `csv.writer`, create a file named `sales.csv` with the following header and data rows:

   ```text
   item,qty,price
   latte,3,4.50
   muffin,2,3.25
   mocha,1,5.00
   ```

   Open the file with `newline=""` and `encoding="utf-8"` when writing.
3. Open `sales.csv` with `csv.reader`.
4. Read and store the header using `next(reader)`.
5. Loop through the remaining rows and calculate:

   ```text
   total_sales = sum of (quantity × price) for all rows
   ```

   CSV values are read as strings, so convert quantity with `int(...)` and price with `float(...)`.
6. Print the header and print the total formatted as money:

   ```text
   Total sales: $25.00
   ```

7. Add a one-sentence **comment** explaining why CSV values must be converted before performing numerical calculations.

### Hints

- `writer.writerows(rows)` writes multiple rows.
- `next(reader)` reads the next row and moves the reader forward.
- `int("3") * float("4.50")` performs numerical multiplication.
- Do not directly print `$25.00`; calculate the total from the rows.


````python
# Meghan Kajihara — Problem 3

import csv

csv_file = "sales.csv"
total_sales = 0.0

csv_contents = [
    ["item","qty","price"],
    ["latte","3","4.50"],
    ["muffin","2","3.25"],
    ["mocha","1","5.00"],
]

with open(csv_file, "w", newline="", encoding="utf-8") as file:
  csv.writer(file).writerows(csv_contents)

with open(csv_file, "r", encoding="utf-8") as file:
  reader = csv.reader(file)
  header = next(reader)

  for item, qty_text, price_text in reader:
    quantity = int(qty_text)
    price = float(price_text)
    row_total = quantity * price
    total_sales += row_total

print(header)
print(f"Total sales: ${total_sales:.2f}")

# CSV values must be converted before performing numerical calculations because csv contents are stored as text.
````

## Problem 4 — Calculations with NumPy *(10 points)*

Use a NumPy array to perform numerical calculations without writing loops.

### What to do

1. Import NumPy using the standard alias: `import numpy as np`.
2. Create the following NumPy array of daily high temperatures in degrees Fahrenheit:

   ```python
   temps = np.array([68, 72, 70, 74, 69, 73, 71])
   ```

3. Print each result with a clear label:

   - mean: `71.0`
   - maximum: `74`
   - minimum: `68`
   - population standard deviation, rounded to two decimal places: `2.0`

4. Use **Boolean indexing** to select temperatures greater than 71. Print the selected values and their count.

   Expected values:

   ```text
   [72 74 73]
   Count: 3
   ```

5. Use one **vectorized expression** to convert all Fahrenheit values to Celsius:

   ```text
   Celsius = (Fahrenheit - 32) × 5 / 9
   ```

   Round the resulting array to one decimal place and print it. Do not use a loop.

   Expected result:

   ```text
   [20.  22.2 21.1 23.3 20.6 22.8 21.7]
   ```

6. Add a one-sentence **comment** naming one advantage of a NumPy array over a Python list for numerical work.

### Hints

- `np.array([...])` creates a NumPy array.
- You may use either NumPy functions, such as `np.mean(temps)`, or equivalent array methods, such as `temps.mean()`.
- `temps > 71` creates a Boolean array.
- `temps[temps > 71]` keeps only values whose Boolean positions are `True`.
- `np.round(array, 1)` rounds each element to one decimal place.


````python
# Meghan Kajihara — Problem 4

import numpy as np

temps = np.array([68, 72, 70, 74, 69, 73, 71])

print("Mean:", temps.mean())
print("Maximum:", temps.max())
print("Minimum:", temps.min())
print("Population Standard Deviation:", round(float(temps.std()), 2))

over_71 = temps[temps > 71]
print(over_71)
print("Count:", len(over_71))

celcius = (temps - 32) * 5 / 9
print(np.round(celcius, 1))

# One advantage of a numpy array over a python list is that they are more efficient at performing calculations on large datasets.
````

## Problem 5 — Save, Load, and Analyze Data with pandas *(10 points)*

Create a pandas DataFrame, save it as a CSV file, load it again, and answer questions about the data.

### What to do

1. Import pandas using the standard alias: `import pandas as pd`.
2. Create a DataFrame from the following dictionary:

   ```python
   data = {
       "name":  ["Maria", "Devon", "Aisha", "Carlos", "Lena"],
       "drink": ["latte", "mocha", "latte", "espresso", "mocha"],
       "spent": [12.50, 9.00, 15.25, 4.50, 9.00],
   }
   ```

3. Store the first DataFrame in a meaningfully named variable such as `customer_data`.
4. Save it as `customers.csv` using `.to_csv(..., index=False)`.
5. Load `customers.csv` with `pd.read_csv(...)` and store the loaded DataFrame in `df`.
6. Print the complete loaded DataFrame.
7. Print each result with a clear label:

   - shape: `(5, 3)`
   - total spent: `$50.25`
   - average spent, rounded to two decimal places: `$10.05`

8. Filter `df` to show only the `name` and `spent` columns for customers who spent **more than $10**. The result should contain Maria and Aisha.
9. Use `df["drink"].value_counts()` to print the number of customers who ordered each drink.
10. Add a one-sentence **comment** explaining the difference between a pandas DataFrame and a pandas Series.

### Hints

- `pd.DataFrame(data)` creates a table from a dictionary.
- `df.to_csv("customers.csv", index=False)` prevents pandas from saving an extra index column.
- `pd.read_csv("customers.csv")` loads a CSV file as a DataFrame.
- `df[df["spent"] > 10][["name", "spent"]]` filters rows and selects columns.
- `value_counts()` may display categories with equal counts in a different order. Correct counts matter more than order.


````python
# Meghan Kajihara — Problem 5

import pandas as pd

data = {
    "name":  ["Maria", "Devon", "Aisha", "Carlos", "Lena"],
    "drink": ["latte", "mocha", "latte", "espresso", "mocha"],
    "spent": [12.50, 9.00, 15.25, 4.50, 9.00],
}

customer_data = pd.DataFrame(data)

customer_data.to_csv("customers.csv", index=False)

df = pd.read_csv("customers.csv")
print(df)

print("Shape:", df.shape)
print(f"Total spent: ${sum(df["spent"])}")
print(f"Average: ${(sum(df["spent"]))/(len(df["spent"])):.2f}")

df[df["spent"] > 10][["name", "spent"]]

df["drink"].value_counts()

# The difference between a pandas dataframe and series is that a dataframe is 2 dimensional while a series is 1 dimensional.
````
