## Setup — Run This First

This cell imports the required libraries and defines `get_soup(url)`.

The helper tries the live website first. If the website cannot be reached, it uses a small backup page with the same HTML structure.

**Do not change this setup cell.**

````python
import requests
from bs4 import BeautifulSoup
import csv
import os
import pandas as pd

# A small backup page with the same structure as Quotes to Scrape.
BACKUP_HTML = """
<html>
<body>
<div class="quote">
    <span class="text">"The world as we have created it is a process of our thinking."</span>
    <small class="author">Albert Einstein</small>
    <div class="tags">
        <a class="tag">change</a>
        <a class="tag">thinking</a>
    </div>
</div>

<div class="quote">
    <span class="text">"It is our choices that show what we truly are."</span>
    <small class="author">J.K. Rowling</small>
    <div class="tags">
        <a class="tag">choices</a>
        <a class="tag">abilities</a>
    </div>
</div>

<div class="quote">
    <span class="text">"Try not to become a person of success, but a person of value."</span>
    <small class="author">Albert Einstein</small>
    <div class="tags">
        <a class="tag">success</a>
        <a class="tag">value</a>
    </div>
</div>
</body>
</html>
"""

def get_soup(url):
    """Download a webpage and return a BeautifulSoup object."""
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        return BeautifulSoup(response.text, "html.parser")
    except requests.RequestException:
        print("Live website unavailable — using the backup page.")
        return BeautifulSoup(BACKUP_HTML, "html.parser")

print("Setup complete. get_soup(url) is ready.")
````

# Part 2 — Extract One Quote  
## 20 points

Use the `soup` object from Part 1.

### Requirements

1. Use `find("div", class_="quote")` to find the first quote block.
2. Extract and print:
   - the quote text

3. Use `.get_text(strip=True)` to remove extra spaces.

Your output should follow this format:

### Reminder

````python
# Name: Meghan Kajihara
# Part 2 — Extract one quote

first_block = soup.find("div", class_="quote")

text = soup.find("span", class_="text")
author = soup.find("small", class_="author")

print("Text:", text.get_text(strip=True))
print("Author:", author.get_text(strip=True))
````

# Part 3 — Extract All Quotes  
## 30 points

Now collect every quote on the page.

### Requirements

1. Use `find_all("div", class_="quote")` to collect all quote blocks.
2. Print the number of quotes found.
3. Create an empty list named `records`.
4. Loop through all quote blocks.
5. For each quote, extract:
   - `text`
   - `author`
   - `tags`
6. Store each quote as a dictionary:

```text
{"author": author, "text": text, "tags": tags}
```

7. Append each dictionary to `records`.
8. Print each author and their list of tags.

### Expected structure

```text
Author Name -> ['tag1', 'tag2']
```

### Reminder

- `find(...)` returns one match.
- `find_all(...)` returns a list of all matches.
- The `tags` value should still be a Python list in this part.

````python
# Name: Meghan Kajihara
# Part 3 — Extract all quotes

all_quotes = soup.find_all("div", class_="quote")
print("Quotes found:", len(all_quotes))

records = []

for quote in all_quotes:
  text = quote.find("span", class_="text").get_text(strip=True)

  author = quote.find("small", class_="author").get_text(strip=True)

  tag_pieces = quote.find_all("a", class_="tag")

  tags = []

  for tag_piece in tag_pieces:
    tag_text = tag_piece.get_text(strip=True)
    tags.append(tag_text)

  record = {
      "author": author,
      "text": text,
      "tags": tags
  }

  records.append(record)

for record in records:
  print(record["author"],"->", record["tags"])
````

````python
# Name: Meghan Kajihara
# Part 4 — Save and analyze the data

with open("quotes.csv", "w", encoding="utf-8") as file:
  writer = csv.DictWriter(file, fieldnames=["author", "text", "tags"])
  writer.writeheader()

  for record in records:
    writer.writerow({
        "author": record["author"],
        "text": record["text"],
        "tags": ";".join(record["tags"])
    })

print("CSV exists?", os.path.exists("quotes.csv"))

df = pd.read_csv("quotes.csv")

print("First 5 rows:")
print(df.head())

print("Shape:", df.shape)
print(df["author"].value_counts())

# The difference between find() and find_all() is that find() returns only the first record and find_all returns all of them.
# One responsible rule for web scraping is check if an API or other public data source is available before scraping.
````
